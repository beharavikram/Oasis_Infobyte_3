```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load the dataset
df = pd.read_excel('/kaggle/input/sentiment-analysis/Sentiment Analysis.xlsx')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Example of preprocessing function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # You might want to add more preprocessing steps here
    return text
```

```python
# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)

# Transform the text data
X = vectorizer.fit_transform(df['Review Text'])
```

```python
# Define features and labels
X = vectorizer.transform(df['Review Text'])
y = df['Sentiment']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train Naive Bayes classifier
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

# Initialize and train Support Vector Machine classifier
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
```

```python
# Predict using Naive Bayes model
y_pred_nb = nb_model.predict(X_test)
print("Naive Bayes Model Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_nb)}")
print(classification_report(y_test, y_pred_nb))
```

```python
# Predict using SVM model
y_pred_svm = svm_model.predict(X_test)
print("SVM Model Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm)}")
print(classification_report(y_test, y_pred_svm))
```
